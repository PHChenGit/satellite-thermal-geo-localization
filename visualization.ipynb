```python
from PIL import Image
import matplotlib.pyplot as plt
import yaml
import os
import numpy as np

folder_config_path = './folder_config.yml'
datasets_folder = './datasets/'
Image.MAX_IMAGE_PIXELS = None
```

```python
# Load folder config
with open(folder_config_path, 'r') as f:
    folder_config = yaml.safe_load(f)
print(folder_config)
```

```python
# Load images
satellite_image_1 = np.array(Image.open(os.path.join(datasets_folder, folder_config['satellite']['name'], folder_config['satellite']['maps'][0])))
satellite_image_2 = np.array(Image.open(os.path.join(datasets_folder, folder_config['satellite']['name'], folder_config['satellite']['maps'][1])))
satellite_image_3 = np.array(Image.open(os.path.join(datasets_folder, folder_config['satellite']['name'], folder_config['satellite']['maps'][3])))
thermal_image_1 = np.array(Image.open(os.path.join(datasets_folder, folder_config['thermalmapping']['name'], folder_config['thermalmapping']['maps'][0])))
thermal_image_2 = np.array(Image.open(os.path.join(datasets_folder, folder_config['thermalmapping']['name'], folder_config['thermalmapping']['maps'][2])))
thermal_image_3 = np.array(Image.open(os.path.join(datasets_folder, folder_config['thermalmapping']['name'], folder_config['thermalmapping']['maps'][4])))
sirmione_image_1 = np.array(Image.open(os.path.join(datasets_folder, folder_config['sirmionemapping']['name'], folder_config['sirmionemapping']['maps'][0])))
foxtech_image_1 = np.array(Image.open(os.path.join(datasets_folder, folder_config['foxtechmapping']['name'], folder_config['foxtechmapping']['maps'][0])))
```

```python
# Crop
satellite_image_crop_1 = satellite_image_1[3000:3500, 2000:5000, :]
satellite_image_crop_2 = satellite_image_2[3000:3500, 2000:5000, :]
satellite_image_crop_3 = satellite_image_3[3000:3500, 2000:5000, :]
thermal_image_crop_1 = thermal_image_1[3000:3500, 2000:5000, :]
thermal_image_crop_2 = thermal_image_2[3000:3500, 2000:5000, :]
thermal_image_crop_3 = thermal_image_3[3000:3500, 2000:5000, :]
sirmione_image_1 = sirmione_image_1[3000:3500, 2000:5000, :]
foxtech_image_1 = foxtech_image_1[3000:3500, 2000:5000, :]
```

```python
# Show
plt.figure(dpi=250)
plt.axis('off')
plt.imshow(satellite_image_crop_1)
plt.figure(dpi=250)
plt.axis('off')
plt.imshow(satellite_image_crop_2)
plt.figure(dpi=250)
plt.axis('off')
plt.imshow(satellite_image_crop_3)
plt.figure(dpi=250)
plt.axis('off')
plt.imshow(thermal_image_crop_1)
plt.figure(dpi=250)
plt.axis('off')
plt.imshow(thermal_image_crop_2)
plt.figure(dpi=250)
plt.axis('off')
plt.imshow(thermal_image_crop_3)
plt.figure(dpi=250)
plt.axis('off')
plt.imshow(sirmione_image_1)
plt.figure(dpi=250)
plt.axis('off')
plt.imshow(foxtech_image_1)
```
